The dataset we are analyzing monitors and tests variables such as gender, ethnicity, parental level of education, lunches, and test preparation to understand how a student’s performance can be impacted. The student’s performance is measured in this dataset by their test scores. 
The variables are as follows: gender (male/female), race/ethnicity (divided into five groups A, B, C, D, and E), parent’s level of education (bachelor's degree, some college, master's degree, associate's degree,- high school), lunch (standard or free/reduced lunch), test preparation (complete or not complete before testing), math, reading and writing scores. 

We chose this dataset because it is extensive with many data points so we can find richer results. 


We are going to run the following analyses on the dataset to decipher what factors are extremely impactful: 
1. Box Plots of scores in relation to test prep and gender 
2. bar graphs of scores related to lunch and parental education 
3. tables depiciting average scores by race and parental education 
4. line and bar graphs of race vs parental education vs score 

There are the initial obvious hypotheses that we are going into the project with, i.e., the higher the level of parent’s education, the higher the student’s test score, completion of test prep implies higher performance, etc. We can map different variables against each other and observe the independence vs dependence of factors. For example, we would not expect a dependence between the race and gender columns, but that might not be the case for race vs level of parent’s education given the historical context in America. It would be meaningful to see which factors directly influence each other or are correlated in influencing the performance of students across math, writing, and reading. 

This data set is exceptionally important because of how relevant it is to the ongoing socio-political conversations. Race and gender are especially interesting factors to observe as we expect them to break harmful stereotypes and emphasize nature vs nurture.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('study_performance.csv')

# Calculate the average math score for each combination of race_ethnicity and parental_level_of_education
avg_math_scores = df.groupby(['race_ethnicity', 'parental_level_of_education'])['math_score'].mean().reset_index()

# Plot the line plot for average math score
plt.figure(figsize=(12, 8))
sns.lineplot(data=avg_math_scores, x='parental_level_of_education', y='math_score', hue='race_ethnicity', marker='o')
plt.title('Average Math Score by Parental Level of Education and Race/Ethnicity')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Math Score')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Race/Ethnicity')
plt.tight_layout()
plt.show()

avg_math_scores = df.groupby(['race_ethnicity', 'parental_level_of_education'])['reading_score'].mean().reset_index()

#similar process for reading and writing scores

# Plot the line plot
plt.figure(figsize=(12, 8))
sns.lineplot(data=avg_math_scores, x='parental_level_of_education', y='reading_score', hue='race_ethnicity', marker='o')
plt.title('Average Reading Score by Parental Level of Education and Race/Ethnicity')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Math Score')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Race/Ethnicity')
plt.tight_layout()
plt.show()

avg_math_scores = df.groupby(['race_ethnicity', 'parental_level_of_education'])['writing_score'].mean().reset_index()

# Plot the line plot
plt.figure(figsize=(12, 8))
sns.lineplot(data=avg_math_scores, x='parental_level_of_education', y='writing_score', hue='race_ethnicity', marker='o')
plt.title('Average Writing Score by Parental Level of Education and Race/Ethnicity')
plt.xlabel('Parental Level of Education')
plt.ylabel('Average Math Score')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Race/Ethnicity')
plt.tight_layout()
plt.show()

#Plotting the distribution of parental level of education by race/ethnicity
plt.figure(figsize=(12, 8))
sns.countplot(data=df, x='race_ethnicity', hue='parental_level_of_education')
plt.title('Distribution of Parental Level of Education by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Parental Level of Education')
plt.tight_layout()
plt.show()

#Plotting the distribution of race/ethnicity
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='race_ethnicity')
plt.title('Distribution of Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

<p style="font-size: 15px;">These plots are quite interesting because we can see that race and parental education are closely related, but parental level of education is not as directly related to scores as observed in the tables. Having a master's degree definitely leads to higher scores in writing and reading, but the impact is not as significant when it comes to math. In fact, students whose parents had a bachelor's degree from Race E did better than those with a master's degree. Also, interestingly, Group A students who performed the worst on math had parents who recieved a master's degree. 

While Group E mostly dominates, Groups B and D seem to alternate and overlap a lot. Even Group C had the highest average among students whose parents have a bachelor's degree in terms of the writing score. Group A tends to score on the lower or lowest end.

Upon looking more closely at the sample size collected from each group, we can also note that Group A has the least number of people, followed by Group E. This is quite an interesting finding, as the data does not have a consistent sample size for each category.

Overall, we can't propose the strongest correlation between parental education and student performance, but it is a pattern that can be observed.
